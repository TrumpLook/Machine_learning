6. Рассмотрим случайную матрицу размером 10х2, представляющую декартовы координаты, преобразуем их в полярные координаты

````python
cartesian = np.random.rand(10, 2) * 10 - 5  # значения от -5 до 5

x = cartesian[:, 0]
y = cartesian[:, 1]

r = np.sqrt(x**2 + y**2)          # Радиус
phi = np.arctan2(y, x)            # Угол (в радианах)

polar = np.column_stack((r, phi))

print("Декартовы координаты:")
print(cartesian)
print("\nПолярные координаты (r, φ):")
print(polar)
````

Классический начальный набор данных - данные пассажиров титаника.

```

Ticket - номер билета

Cabin - каюта
Embarked - причал отбытия. C = Cherbourg Q = Queenstown S = Southampton
```

````python
import pandas as pd
````

````python
titanic = pd.read_csv('data.csv')
````

1. Узнайте сколько было на борту человек

````python
total_passengers = len(titanic)
print(f"Всего пассажиров на Титанике: {total_passengers}")
````

2. Какой медианный и средний возраста пассажиров

````python
print(titanic['Age'].isnull().sum(), "пропущенных значений в Age")
mean_age = titanic['Age'].mean()
median_age = titanic['Age'].median()
print(f"Средний возраст: {mean_age:.1f} лет")
print(f"Медианный возраст: {median_age:.1f} лет")
````

3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
def age_group(age):
    if pd.isnull(age):
        return 'Unknown'
    return 'Child' if age < 16 else 'Adult'

titanic['Age_Group'] = titanic['Age'].apply(age_group)
survival_rate = titanic.groupby('Age_Group')['Survived'].mean() * 100
print("Процент выживаемости:")
print(survival_rate[['Child', 'Adult']]) 
print("\nКоличество пассажиров в каждой группе:")
print(titanic['Age_Group'].value_counts())
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
print(titanic.groupby(['Pclass', 'Sex', 'Age_Group'])['Survived'].mean().unstack() * 100)
````

5. Зависит ли выживаемость от класса обслуживания?

````python
survival_by_class = titanic.groupby('Pclass')['Survived'].mean() * 100
print("Выживаемость по классам (%):")
print(survival_by_class)
````

6. Посчитайте средний возраст умерших женщин и мужчин

````python
dead_women = titanic[(titanic['Survived'] == 0) & (titanic['Sex'] == 'female')]
dead_men = titanic[(titanic['Survived'] == 0) & (titanic['Sex'] == 'male')]
avg_age_dead_women = dead_women['Age'].mean()
avg_age_dead_men = dead_men['Age'].mean()
avg_age_dead_women, avg_age_dead_men
````

7. Различается ли процент выживаемости пассажиров с хотя бы одним родственником на борту и одиночек?

````python
titanic['HasFamily'] = (titanic['SibSp'] + titanic['Parch']) > 0
survival_rates = titanic.groupby('HasFamily')['Survived'].mean() * 100
survival_rates
````

8. Различается ли средняя стоимость билета у умерших и выживших пассажиров?

````python
fare_by_survival = titanic.groupby('Survived')['Fare'].mean()
print(fare_by_survival)
````

9. Выведите максимальный и минимальный возраст выживших пассажиров во 2 классе обслуживания

````python
survived_class2 = titanic[(titanic['Survived'] == 1) & (titanic['Pclass'] == 2)]
min_age = survived_class2['Age'].min()
max_age = survived_class2['Age'].max()
print(f"Минимальный возраст: {min_age}")
print(f"Максимальный возраст: {max_age}")
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
men = titanic[(titanic['Sex'] == 'male') & (titanic['Age'].notna())]
young_men = men[men['Age'] <= 18]
adult_men = men[men['Age'] > 18]
young_survival_rate = young_men['Survived'].mean() * 100
adult_survival_rate = adult_men['Survived'].mean() * 100
print(f"Процент выживших молодых мужчин (до 18): {young_survival_rate:.2f}%")
print(f"Процент выживших взрослых мужчин: {adult_survival_rate:.2f}%")
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Постройте гистограмму распределения возростов пассажиров

````python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
titanic = pd.read_csv('data.csv')

# Удаление пропущенных значений возраста
ages = titanic['Age'].dropna()

# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=30, color='skyblue', edgecolor='black')
plt.title('Распределение возрастов пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.grid(True)
plt.show()
````

2. Постройте гистограммы распределения цен для пассажиров разных классов

````python
plt.figure(figsize=(10, 6))
for pclass in titanic['Pclass'].unique():
    subset = titanic[titanic['Pclass'] == pclass]
    plt.hist(subset['Fare'], bins=30, alpha=0.5, label=f'Класс {pclass}')

plt.title('Распределение стоимости проезда для пассажиров разных классов')
plt.xlabel('Стоимость проезда (Fare)')
plt.ylabel('Частота')
plt.legend(title='Класс обслуживания')
plt.show()
````

3. Постройте зависимость цены от возраста пассажира

````python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5, c='blue', edgecolors='k', s=50)
plt.title('Зависимость стоимости проезда от возраста пассажира')
plt.xlabel('Возраст пассажира (Age)')
plt.ylabel('Стоимость проезда (Fare)')
plt.show()
````

4. Постройте box plot отображающий распределение цен на билеты в разных классах

````python
plt.figure(figsize=(10, 6))
titanic.boxplot(column='Fare', by='Pclass', vert=False)
plt.title('Распределение цен на билеты в разных классах')
plt.suptitle('') 
plt.xlabel('Стоимость проезда (Fare)')
plt.ylabel('Класс обслуживания (Pclass)')
plt.show()
````

5. Визуализируйте распределение долей выживших среди мужчин, женщин и детей(до 16 лет)

````python
titanic['Category'] = titanic['Sex']
titanic.loc[titanic['Age'] < 16, 'Category'] = 'Child'
survival_rate = titanic.groupby('Category')['Survived'].mean()
plt.figure(figsize=(8, 6))
survival_rate.plot(kind='bar', color=['blue', 'pink', 'green'])
plt.title('Распределение долей выживших среди мужчин, женщин и детей')
plt.xlabel('Категория')
plt.ylabel('Доля выживших')
plt.xticks(rotation=0)
plt.show()
````

6. Сравните плотности распределения возростов выживших и умерших пассажиров, сделайте выводы

````python
survived = titanic[titanic['Survived'] == 1]
died = titanic[titanic['Survived'] == 0]
plt.figure(figsize=(10, 6))
plt.hist(survived['Age'].dropna(), bins=30, density=True, alpha=0.5, label='Выжившие', color='green')
plt.hist(died['Age'].dropna(), bins=30, density=True, alpha=0.5, label='Умершие', color='red')
plt.title('Сравнение плотности распределения возрастов выживших и умерших пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.show()
````
